Avoir les bans de tout les champions
write_multiple_ban() ou print_multiple_bans()

```python

def get_ban(champion, db='database_top_team_2024', win=False):
    """
    take a champion as entry
    return a dict with for each champion the number of time he was banned
    """
    champ_bans = {}
    for champ in champion_list:
        champ_bans[champ] = 0
    
    for i in range(1,6):
        if win:
            res = do_sql(f"SELECT ban{i} FROM {db} WHERE champion = \"{champion}\" AND result = '1'", cur).fetchall()
        else:
            res = do_sql(f"SELECT ban{i} FROM {db} WHERE champion = \"{champion}\"", cur).fetchall()
        for champ in res:
            if champ[0] != None:
                champ = champ[0]
                champ_bans[champ] += 1

    return champ_bans


def print_ban(dictio):
    """
    simply print from a dict from get_ban(). good for copy/paste into an excel file
    """
    for champion, nb_of_bans in dictio.items():
        # print(f"{champion}, {nb_of_bans}")
        print(nb_of_bans)

def print_multiple_bans(list_of_dictio):
    """
    print for_each champions of champion_list the number of bans. good for copy/paste into an excel file
    """
    for champ in champion_list:
        for dictio in list_of_dictio:
            print(dictio[champ], end=",")
        print()


def write_multiple_bans(list_of_dictio):
    """
    same as print_multiple_bans() but write it in a csv file instead of printing it
    """
    champion_list2 = deepcopy(champion_list)
    champion_list2.insert(0, None)
    x = 1
    result_list = [champion_list2, []]
    # print("champion_list2", champion_list2)
    # print("result_list", result_list)
    # print("list_of_dictio", list_of_dictio)

    for champ in champion_list:
        result_list[x].append(champ)
        for dictio in list_of_dictio:
            result_list[x].append(dictio[champ])
        result_list.append([])
        x +=1
    
    with open("bans.csv", mode='w', newline='') as fichier_csv:
        writer = csv.writer(fichier_csv)
        writer.writerows(result_list)


def create_list_of_bans():
    """
    create a dict with for each champions their bans
    this list is used in the others functions
    """
    list_of_dictio = []
    for i in range(len(champion_list)):
        bans = get_ban(champion_list[i])
        list_of_dictio.append(bans)
    return list_of_dictio

write_multiple_bans(list_of_dictio=create_list_of_bans())
```

```python
con = create_connection(path=r'D:\code\data_analyse\2024-total-database.db')
cur = con.cursor()

def get_game_id():
    res = do_sql(f'SELECT gameid FROM {db["database"]}', cur).fetchall()
    list_of_game_id = []
    for gameid in res:
        if not gameid[0] in list_of_game_id:
            list_of_game_id.append(gameid[0])
    return list_of_game_id

def get_draft():
    """
    algo demandant en ressource
    retourne toutes les drafts de la forme: [[champ1, champ2, champ3, champ4, champ5],[...]]
    """
    list_of_game_id = get_game_id()
    print("nb de games", len(list_of_game_id))
    drafts = []
    # for i in range(5):
    #     gameid = list_of_game_id[i]
    for gameid in list_of_game_id:
        for side in ['Blue', 'Red']:
            draft = do_sql(f"SELECT champion FROM {db['database']} WHERE gameid = '{gameid}' AND side='{side}'", cur).fetchall()
            result = do_sql(f"SELECT result FROM {db['database']} WHERE gameid = '{gameid}' AND side='{side}'", cur).fetchone()
            clean_draft = []
            for champ in draft:
                if champ[0] != None:
                    clean_draft.append(champ[0])
            clean_draft.append(int(result[0]))
            drafts.append(clean_draft)
    
    draft2 = {}
    for draft in drafts:
        if not str(draft) in draft2.keys():
            draft2[str(draft)] = 1
        else:
            draft2[str(draft)] += 1
    # drafts = [key[2::] for key, repet in draft2.items()]
    final_drafts = []
    for el, result in draft2.items():
        el = eval(el)
        el.append(result)
        final_drafts.append(el)

    return final_drafts

def write_drafts():
    drafts = get_draft()

    with open(f'draft_{db["database"]}.csv', 'w', newline='') as ficher_csv:
        csv_writer = csv.writer(ficher_csv)
        fieldnames = ['top','jungle','mid', 'adc', 'support', 'result', 'count']
        csv_writer.writerows([fieldnames])
        csv_writer.writerows(drafts)
    
def get_drafts_from_csv():
    drafts = []

    with open(r'draft.csv', newline='') as csvfile:
        lecteur_csv = csv.reader(csvfile, delimiter=',')
        for ligne in lecteur_csv:
            drafts.append(ligne)

    return drafts    


def get_draft_for_champion_count(champion):
    res = do_sql(f'SELECT top, jungle, mid, adc, support FROM {db["draft"]} WHERE top = "{champion}" OR jungle = "{champion}" OR mid = "{champion}" OR adc = "{champion}" OR support = "{champion}"', cur).fetchall()
    champions_in_draft = []
    for draft in res:
        for champions in draft:
            if champions != champion:
                champions_in_draft.append(champions)

    champ_in_draft_count = {}
    for champ in champion_list:
        champ_in_draft_count[champ] = 0
    
    for champions in champions_in_draft:
        champ_in_draft_count[champions] += 1
    
    print(champ_in_draft_count)
    return champ_in_draft_count

def write_champions_in_draft_count():
    """
    semble écrire pour chaque champ le nb de games faites avec un autre oui
    """
    list_of_dictio = []
    for i in range(len(champion_list)):
        champ_in_draft_count = get_draft_for_champion_count(champion_list[i])
        list_of_dictio.append(champ_in_draft_count)

    champion_list2 = deepcopy(champion_list)
    champion_list2.insert(0, None)
    x = 1
    result_list = [champion_list2, []]
    # print("champion_list2", champion_list2)
    # print("result_list", result_list)
    # print("list_of_dictio", list_of_dictio)

    for champ in champion_list:
        result_list[x].append(champ)
        for dictio in list_of_dictio:
            result_list[x].append(dictio[champ])
        result_list.append([])
        x +=1
    
    print('result_list',result_list)
    with open("draft_count.csv", mode='w', newline='') as fichier_csv:
        writer = csv.writer(fichier_csv)
        writer.writerows(result_list)

# write_champions_in_draft_count()
write_drafts()
```

```python
def create_sql_file_from_csv(input_file, table_name):

    # Ouverture du fichier CSV en mode lecture
    with open(input_file, 'r') as csvfile:
        # Lecture du fichier CSV
        csvreader = csv.reader(csvfile)
        # print("csv_reader1", csvreader[0])
        # Ignorer l'en-tête du CSV
        fieldnames = next(csvreader)
        fieldnames_2 = ""
        fieldnames_3 = ""
        for fields in fieldnames:
            fieldnames_2 += f"{fields} VARCHAR(255) NULL, "
            fieldnames_3 += f"{fields}, "
        fieldnames_2 = fieldnames_2[:-2]
        fieldnames_3 = fieldnames_3[:-2]
        # Afficher la commande SQL de création de la table
        con.commit()
        do_sql(f"CREATE TABLE {table_name} ({fieldnames_2});")
        # Parcourir les lignes du CSV et générer les commandes SQL
        for row in csvreader:
            result = ""
            for el in row:
                result += f'"{el}", '
            result = result[:-2]
            # adc = row[0]
            # supp = row[1]
            # count = row[2]
            
            # Générer la commande SQL d'insertion
            do_sql(f"INSERT INTO {table_name} ({fieldnames_3}) VALUES ({result});")
    con.commit()

create_sql_file_from_csv(r'draft_database_2023.csv', r"draft2_database_2023")
```
